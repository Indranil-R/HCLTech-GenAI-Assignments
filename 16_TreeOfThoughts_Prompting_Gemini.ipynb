````python
import os
from langchain.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains import LLMChain
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda
from IPython.display import display, Markdown
````

````python
from google.colab import userdata
os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")
````

````python

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-preview-04-17", temperature=0,cache=False)
parser = StrOutputParser()

brainstorm_prompt = PromptTemplate(
    input_variables=["question"],
    template="""


I have a question related to {question}. Could you brainstorm three distinct solutions?

""",
)
chain1 = brainstorm_prompt | llm | parser


evaluation_prompt = PromptTemplate(
    input_variables=["solutions"],
    template="""

For each of the three proposed solutions, evaluate their potential. Consider their pros and cons, initial effort needed, implementation difficulty, potential challenges, and the expected outcomes. Assign a probability of success and a confidence level to each option based on these factors.

{solutions}

""",
)
chain2 = evaluation_prompt | llm | parser

deep_dive_prompt = PromptTemplate(
    input_variables=["review"],
    template="""

For each solution, deepen the thought process. Generate potential scenarios, strategies for implementation, any necessary partnerships or resources, and how potential obstacles might be overcome. Also, consider any potential unexpected outcomes and how they might be handled.

{review}

""",
)
chain3 = deep_dive_prompt | llm | parser

ranking_prompt = PromptTemplate(
    input_variables=["deepen_thought_process"],
    template="""

Based on the evaluations and scenarios, rank the solutions in order of promise. Provide a justification for each ranking and offer any final thoughts or considerations for each solution.

{deepen_thought_process}

""",
)
chain4 = ranking_prompt | llm | parser
````

````python
from langchain.chains import SequentialChain

def print_logs(label,content):
  display(Markdown(f"\n### {label}:\n{content}"))


def append_log(label):
    # return RunnableLambda(lambda x: (display(Markdown(f"\n### [{label}]:\n\n```markdown\n{x}\n```")),x)[1])
    return RunnableLambda(lambda x: (print_logs(label,content=x),x)[1])


full_chain = (
    chain1
    | append_log("Brainstormed Solutions")
    | chain2
    | append_log("Evaluated Solutions")
    | chain3
    | append_log("Deep Dive")
    | chain4
)

result = full_chain.invoke({"question": "India becoming a global superpower"})
````
